```python
psi_file = "PSI_WT_vs_KO_filtered.csv"
exon_file = "optional_exons.bed"
output_file = "delta_PSI_exons.bed"

MIN_DELTA = 0.2

exon_dict = {}

with open(exon_file) as f:
    for line in f:
        parts = line.strip().split("\t")
        chr = parts[0]
        start = parts[1]
        end = parts[2]
        gene = parts[3]
        exon_dict[(gene, start, end)] = chr

with open(psi_file) as infile, open(output_file, "w") as outfile:

    header = infile.readline()

    for line in infile:
        parts = line.strip().split(",")

        gene = parts[0]
        start = parts[1]
        end = parts[2]
        delta = float(parts[5])

        if abs(delta) >= MIN_DELTA:
            chr = exon_dict.get((gene, start, end))
            if chr:
                outfile.write(f"{chr}\t{start}\t{end}\t{gene}\t{delta}\n")

print("delta_PSI_exons.bed erstellt.")
```
