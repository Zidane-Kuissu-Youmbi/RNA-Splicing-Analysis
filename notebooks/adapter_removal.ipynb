```python
def adapter_entfern(input, output, adapter):
    with open(input) as infile, open(output, "w") as outfile:
        while True:
            header = infile.readline()
            if not header:
                break
            seq = infile.readline().rstrip("\n")
            plus = infile.readline()
            qual = infile.readline().rstrip("\n")

            pos = seq.find(adapter)

            if pos != -1:
                seq = seq[:pos]
                qual = qual[:pos]

            if len(seq) == len(qual) and len(seq) > 0:
                outfile.write(header)
                outfile.write(seq + "\n")
                outfile.write(plus)
                outfile.write(qual + "\n")   
adapter_entfern("AGSP7_KO_RIP-Seq.fastq", "KO_ohneAdapter.fastq", "AGATCGGAAGAGCACACGTCTGAACTCCAGTCA")
adapter_entfern("AGSP7_WT_RIP-Seq.fastq", "WT_ohneAdapter.fastq", "AGATCGGAAGAGCACACGTCTGAACTCCAGTCA")
```
